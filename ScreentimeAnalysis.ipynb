```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

Here we have a sample dataset composed of screentime data for a given user across a series of dates. In this data analysis, I will sort the data and plot it in various graphs to display different aspects of the data.

```python
df = pd.read_csv('Screentime.csv')
df = df.set_index('App')
df
```

```python
hours = sns.scatterplot(data=df,x='Date',y='Usage (hours)',hue='Usage (hours)').set_title('Hours Spent on Apps')
hours


plt.xlabel('Date')
plt.xticks(rotation=90, fontsize = 'xx-small')

plt.ylabel('Usage (hours)')
plt.show()
```

This scatter plot depicts the average hours spent on apps over the series of dates. From this data, we can see that the data tends to range anywhere from 3-10 hours a day. 15 hours a day and 17 hours a day are the two outliers within this dataset.

```python
sns.barplot(x='Date', y='Notifications', data=df).set_title('Number of Notifications a Day')
plt.xticks(rotation=90, fontsize = 'xx-small')
```

This bar plot depicts the number of notifications received a day from apps over a series of dates. The further the dates go, the notification  number appears to have an increase. Notifications reach as high as 400 in Oct 23 and as low as 2 in early October.

```python
df.plot(
    title = '# of Times Apps were Opened',
    kind = 'line',
    x='Date', 
    y='Times opened',
)
```

This line plot depicts the number of times apps were opened over a series of dates. There are some significant spikes around late September/early October and in late October. Numbers are their lowest in early November and  mid September.

```python
hours = sns.scatterplot(data=df,x='App',y='Usage (hours)',hue='Usage (hours)').set_title('Most Used App')
hours


plt.xlabel('App')


plt.ylabel('Usage (hours)')
plt.show()
```

This scatter plot depicts the app usage. All of the apps tend to have a similar amount of usage however TikTok has the most usage out of all of them. The average app usage ranges anywhere from 2 to 10 hours a day. The two outliers is Twitter which has one day with a usage of about 16 hours and TikTok which has one day with a usage of about 17.5 hours.
